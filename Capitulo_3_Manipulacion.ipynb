# 1. Agregaci贸n de datos
---


En el contexto del an谩lisis de datos, *agregar* data significa realizar c谩lculos que permitan resumir informaci贸n, abstrayendo conocimiento de data que se encuentra en estado bruto. 

>  Se trata del ejercicio que hacemos cada vez que utilizamos una tabla din谩mica (pivot table) en Excel


Al realizar una agregaci贸n debemos establecer *el tipo de c谩lculo que buscamos realizar*. Para ello, utilizamos funciones espec铆ficas: *funciones de agregaci贸n*. 

*Funciones de agregaci贸n*

| Funci贸n     | Descripci贸n                     |
|-------------|---------------------------------|
| count       | Contar filas no nulas           |
| nunique     | Contar valores 煤nicos           |
| sum         | Suma de valores                 |
| mean        | Media aritm茅tica                |
| median      | Mediana                         |
| mode        | Moda                            |
| min         | Valor m铆nimo                    |
| max         | Valor m谩ximo                    |

````
import pandas as pd
````

````
# importamos df
df_jamesbond = pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/jamesbond.csv",index_col="Film")
df_jamesbond.dtypes
````

### Agregaciones b谩sicas

Comenzaremos realizando agregaciones simplemente utilizando funciones de aggregaci贸n sobre dataframes.

```
df.agg_func()
```


````
# aplicamos una funci贸n de agregaci贸n sobre el dataframe
df_jamesbond.max()
# obtenemos los valores max para cada una de las variables
````

````
# seleccionamos una variable y aplicamos funci贸n de agregaci贸n
df_jamesbond.Budget.sum().round()
# anidamos un .round() para redondear decimales
````

````
# seleccionamos una variable y aplicamos funci贸n de agregaci贸n
df_jamesbond.Actor.nunique()
# existen 7 actores de James Bond (煤nicos) en el dataframe
````

### El m茅todo .agg()

Pandas incluye un m茅todo dedicado a agregar data.

```
df.agg({'column':'agg_func'})
```
>  Utilizamos un *dictionary* para establacer sobre qu茅 variable realizar qu茅 operaci贸n

````
# agregamos data utilizando .agg()
df_jamesbond.agg({'Actor':"nunique"})
````

````
# agregamos la columna "box office", aplicando funciones min(), max() y mean()
df_jamesbond.agg({'Box Office':["min","max","mean"]})
# en este caso, hemos utilizado m谩s de una funci贸n de agregaci贸n sobre la misma variable
````

````
# agregamos m谩s de una columna a帽adiendo m煤ltiples elementos en nuestro dictionary
df_jamesbond.agg({
    'Budget':["min","max","median"],           
    'Bond Actor Salary':["min","max","median"]   
    })
````

### Agregaciones agrupadas

Al realizar agregaciones, frecuentemente agrupamos observaciones en una o m谩s variables categ贸ricas. 

El m茅todo `.groupby()` permite agrupar observaciones para posteriormente agregar seg煤n grupos.

```
object.groupby("column")
```

````
# agregamos dos m茅tricas, seg煤n una variable agrupada
df_jamesbond.groupby("Actor")[["Bond Actor Salary","Box Office"]].max()
````

````
# repetimos ejercicio utilizando la funci贸n .agg()
df_jamesbond.groupby("Actor").agg(
    {"Bond Actor Salary":"max","Box Office":"max"}
)
````

````
# agregamos salario m谩ximo y media de box office, seg煤n actor
## cambiamos los nombres de las variables agregadas
df = df_jamesbond.groupby("Actor").agg(
    {"Bond Actor Salary":"max","Box Office":"mean"}
).sort_values("Bond Actor Salary", ascending=False)

## muchas veces, los nombres que obtenemos por defecto no son representativos de la informaci贸n que almacenan
df = df.rename(columns={'Bond Actor Salary':'total_bond_salary','Box Office':'total_box_office'})

df # invocamos el objeto
````

````
# agregamos m煤ltiples m茅tricas utilizando distintas funciones
## al momento de ordenar, debemos utilizar un tuple para poder especificar (i) variable y (ii) funci贸n
df_jamesbond.groupby("Actor").agg(
    {"Bond Actor Salary":["max","sum","mean","size"]
    ,"Budget":["max","min"]}
).sort_values(("Bond Actor Salary","max"),ascending=False) # al ordenar el df, debemos utilizar un "tuple" para definir criterio
````

````
# podemos utilizar m谩s de una variable al agregar
## en tal caso, las observaciones se agrupar谩n seg煤n estos dos campo combinados
df_jamesbond.groupby(["Director","Actor"]).agg(
    {"Box Office":"median"}
)
````

# 2. Trabajar con datos hist贸ricos
---
Python no reconoce las fechas como tipo de variable de forma nativa. Para poder operar con ellos, debemos importar librer铆as espec铆ficas.

````
import pandas as pd
````

### Operando con fechas en un DataFrame


````
# importamos df
df_amzn = pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/amzn_stocks.csv")
df_amzn.dtypes
````

````
# usamos la funci贸n pd.to_datetime() para convertir una columna en tipo objeto
df_amzn['Date'] = pd.to_datetime(df_amzn['Date'])
df_amzn.dtypes
````

````
# en caso de que Pandas no identifique autom谩ticamente el formato de fecha, podemos especificarlo editando un par谩metro opcional
df_amzn['Date'] = pd.to_datetime(df_amzn['Date'],format='%Y-%m-%d')
df_amzn.dtypes
````

````
# creamos una nueva columna con el a帽o
df_amzn['year'] = df_amzn['Date'].dt.year
````

````
# creamos una nueva columna con el a帽o
df_amzn['month'] = df_amzn['Date'].dt.month
````

````
# creamos una nueva columna con el nombre de d铆a
df_amzn['day_name'] = df_amzn['Date'].dt.day_name()
````

````
# creamos una nueva columna con el n煤mero de semana
df_amzn['week_num'] = df_amzn['Date'].dt.isocalendar().week
````

````
# visualizamos el objeto
df_amzn.head()
````

### Agregando data hist贸rica
Podemos agregar data agrupando seg煤n componentes de fecha que hayamos calculado, utiliz谩ndolos como variables categ贸ricas.


````
# utilizamos los componentes "a帽o" y "n煤mero de semana" para agrupar nuestra agregaci贸n
# calculamos la media semanal
df_amzn.groupby(['year','week_num']).agg({'Open':'mean'})
````

````
# calculamos la media anual
output = df_amzn.groupby('year').agg({'Open':'mean'})
output
````

El m茅todo `.shift()` permite tomar el valor de X filas anteriores.

````
# utilizamos el m茅todo .shift() para crear una nueva columna que capte el valor del a帽o anterior
output['open_1YA'] = output['Open'].shift(1)
output
````

````
# calculamos crecimiento anual (%)
output['yoy_growth'] = (output['Open'] / output['open_1YA']) - 1
output['yoy_growth'] = output['yoy_growth'].round(2)
output
````

# 3. Unir tablas de datos
---
Con frecuencia nos vemos obligados a unir dos o m谩s DataFrames. 

Podemos distinguir entre distintas operaciones:
* Unir tablas por filas; llamamos estas operaciones *unions*
* Unir por columnas, seg煤n alg煤n campo en com煤n; llamamos estas operaciones *joins*
* Simplemente unir por columnas, sin ning煤n campo en com煤n



````
# importamos df
df_jamesbond = pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/jamesbond.csv")
df_jamesbond.dtypes
````

### Unir tablas por filas

Uniremos dos dataframes utilizando la funci贸n `pd.concat()`. Nota: los datasets deben contener las mismas columnas.

Unions

````
pd.concat([df_jamesbond.head(2),df_jamesbond.tail(2)])
````

### Unir tablas por columnas, seg煤n campos en com煤n

La funci贸n `.merge()` permite instrumentar *joins*; es decir, unir dos DataFrames seg煤n columnas utilizando campo(s) en com煤n. 

>  El m茅todo `.merge()` permite instrumentar distintos tipos jode joins (left, right, inner, full)

```

```

>  En caso de que los *keys* (columnas que usamos para unir) tengan nombres distintos en cada tabla, podemos editar los par谩metros `left_on` y `right_on`

```
df_1.merge(df_2,how='inner',left_on='key',right_on='key')
```

````
# importamos dataframe con data de albumes
df_albums = pd.read_csv('https://data-wizards.s3.amazonaws.com/datasets/rolling_stones_top_metal_albums.csv')
df_albums.dtypes
````

````
# importamos dataframe con data de canciones
df_songs = pd.read_csv('https://data-wizards.s3.amazonaws.com/datasets/rolling_stones_top_metal_songs.csv')
df_songs.dtypes
````

````
df_songs[['Song','Artist','AlbumID_Rank']].merge(
    df_albums[['AlbumID_Rank','Album','Release_Year']]  # df secundario
    ,how='left'                                         # definimos el criterio en el join, en este caso elegimos "left"
    ,left_on='AlbumID_Rank'                             # definimos la key (columna) en la tabla principal 
    ,right_on='AlbumID_Rank'                            # key en nuestra tabla secundaria
)
````

# 4. Pivotar tablas de datos
---
En ocasiones, nos encontamos con tablas de datos que estructuran la informaci贸n de forma inadecuada, dificultando la exploraci贸n y manipulaci贸n. Debemos mantener la unidad de observaci贸n lo m谩s simple posible, de forma tal de favorecer la interpretaci贸n de la tabla de datos.


### De formato *long* a *wide*

El m茅todo *pivot* nos permite separar una variable en m煤ltiples columnas, convirtiendo una tabla *long* en *wide*.

```
df.pivot(
  index= [] # columnas que no deseamos modificar
  ,columns=[] # columnas que deseamos expandir
  ,values='col_name' # nombre m茅trica
)
```

````
# importamos DataFrame
df_dummy_long_table = pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/dummy_long_table.csv")
df_dummy_long_table
````

````
# utilizamos el m茅todo .pivot() para convertir tabla "long" en "wide"
df_dummy_long_table.pivot(index='country',columns=['year','type'],values='count')
````

### De formato *wide* a *long*

El m茅todo *melt* nos permite agrupar m煤ltiples variables, reduciendo el n煤mero de columnas y aumentando la cantidad de filas.

```
df.melt(
  id_vars = [] # columnas que no deseamos modificar
  ,value_vars = [] # columnas que deseamos agrupar
  ,var_name = 'col_name' # nombre variable categ贸rica
  ,value_name = 'col_name' # nombre m茅trica
)
```



````
# importamos tabla de datos
df_dummy_wide_table = pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/dummy_wide_table.csv")
df_dummy_wide_table
````

````
# utilizamos el m茅todo .melt() para convertir df "wide" a "long"
df_dummy_wide_table.melt(var_name='country',value_name='population')
````

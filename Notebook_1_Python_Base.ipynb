*   *Lists*
*   *Dictionaries*

> üëâ Por fuera de estas cuatro estructuras de datos b√°sicas, existen estructuras adicionales que han sido incorporadas por medio de librer√≠as, es decir, funcionalidades que no son parte del c√≥digo fuente de Python. ### Lists
Los *lists* son estructuras *indexadas* que permiten almacenar elementos de cualquier tipo. Se trata de la estructura m√°s sencilla y utilizada, dentro de aquellas disponibles en Python base. 

> üëâ Que una estructura sea *indexada* significa que Python no solamente almacena la informaci√≥n que all√≠ depositemos, sino tambi√©n *la posici√≥n espec√≠fica que ocupa cada elemento dentro de la estructura*  

> ‚ö†Ô∏è Python indexa al cero. Esto significa que el 0 es la primer posici√≥n para este lenguaje de programaci√≥n  

Para declarar un list utilizamos la siguiente sintaxis

```
my_list = [1,2,3]
```



````
# declaramos un list
my_list = [1,"a",3,False,5]
my_list
````

````
# corroboramos que nuestro objeto almacena un list
type(my_list)
````

````
# los lists pueden incluso contener otras estructuras como elementos
my_list_v2 = [100,"cien",my_list]
my_list_v2
````

````
# la caracter√≠stica principal de un list es que los elementos almacenados se encuentran indexados
# esto significa que cada elemento en un list contiene un n√∫mero √∫nico que nos permite invocarlo
# nota importante: Python indexa al 0 (el 0 es el primer valor)
my_list_v2[0]
````

````
# otra ventaja, es que podemos modificar un elemento, simplemente conociendo su posici√≥n
# modificamos un elemento dentro del list
my_list_v2[0] = 101
my_list_v2
````

### Dictionaries
En ocasiones, nos interesa poder invocar elementos en nuestra estructura a partir de valores que hayamos asignado y no seg√∫n su posici√≥n.

Los *dictionaries* son estructuras que permiten almacenar elementos bajo una l√≥gica *key-pair*. Se trata de estructuras no indexadas, donde los elementos all√≠ depositados son invocados seg√∫n un valor *key* asignado.


> ‚ö†Ô∏è Los *dictionaries* no admiten valores *key* duplicados y los elementos almacenados no mantienen un orden ni se encuentran indexados; los valores (*pair*) se invocan utilizando el *key*.

Para declarar un dictionary utilizamos la siguiente sintaxis

```
my_dic = {
  key_1:value_1,
  key_2:value_2,
  key_3:value_3
}
```


````
# declaramos dic
my_dic = {
  "code": "Python",
  "lecturer": 'Juan Martin Bellido',
  "is_practical": True,
  "number_students":3,
  "student_names":["Maria","Juan","Jose"] # utilizamos un list aqu√≠, ya que estamos proporcionando m√°s de un elemento
}

# invocamos dic
my_dic
````

````
# corroboramos que nuestro objeto almacena un dic
type(my_dic)
````

````
# invocamos elemento en dic utilizando su valor "key"
my_dic['lecturer']
````

````
# reemplazamos un elemento almacenado
my_dic['student_names'] = ["Maria","Juan","Jose","Pepe"]
my_dic
````

````
# podemos utilizar el m√©todo .keys() para obtener los valores key en nuestro dic
# un "m√©todo" es similar a una funci√≥n, ver secci√≥n 4 para m√°s detalles
my_dic.keys()
````

# 3. Funciones vs. m√©todos
---
Una caracter√≠stica curiosa de Python es la distinci√≥n entre *funciones* y *m√©todos*. En un sentido amplio, podr√≠amos hablar de ambas como "funciones"; se trata de c√≥digo que invocamos para realizar una tarea espec√≠fica. 

Utilicemos como ejemplo la funci√≥n *type()*, previamente utilizada. Detr√°s de esta funci√≥n hay c√≥digo, al cual a priori no podemos acceder (y francamente no nos interesa hacerlo). Invocamos esta funci√≥n porque conocemos la tarea espec√≠fica que realiza. Por otro lado, para poder actuar, la funci√≥n espera un input, en este caso el nombre de un objeto declarado en nuestro entorno.

Sintaxis gen√©rica de una funci√≥n (propiamente dicha)

```
function_name(parameter_1, parameter_2, ...)
```


*¬øQu√© son los m√©todos?*

Los m√©todos son funciones que est√°n *asociadas a una clase o tipo de informaci√≥n espec√≠fica*. Mientas las funciones (propiamente dichas) son amplias y permiten ser utilizadas con cualquier tipo de objeto, los m√©todos √∫nicamente operan con el tipo de objeto para el cual han sido dise√±ados.

La sintaxis en un m√©todo cambia

```
object.method_name(parameter_1, parameter_2, ...)
```

> ‚ö†Ô∏è En un m√©todo, a diferencia de una funci√≥n, comenzamos siempre con el nombre del objeto

Como ejemplo de un m√©todo, hemos presentado en la secci√≥n anterior el m√©todo *.keys()*. En este caso, hablamos de un m√©todo espec√≠fico para *dictionaries* y que no tendr√≠a sentido para otro tipo de estructuras.

### Expresiones condicionales

Una *expresi√≥n condicional* es una serie de pruebas l√≥gicas con respuestas pre definidas en caso de que la condici√≥n se cumpla.

> üëâ Para realizar prueba l√≥gicas (testear condiciones) utilizamos siempre operadores de comparaci√≥n

Sintaxis b√°sica
```
if (cond):
  respuesta si la condici√≥n se cumple 
```

Sintaxis ampliada
```
if (cond):
  respuesta si la condici√≥n se cumple 
elif (cond):
  respuesta si la condici√≥n se cumple 
...
else: 
  respuesta si ninguna de las condiciones anteriores se cumple
```

> ‚ö†Ô∏è En caso de tener m√∫ltiples condiciones, estas se ir√°n testeando en orden. La primera condici√≥n en cumplirse activar√° la respuesta

````
# creamos una condici√≥n para comprobar si el objeto contiene un n√∫mero positivo
obj = 10 # creamos un objeto

if obj > 0:                         # condici√≥n 1: testear si el objeto es positivo
  print("number is positive")       # print() despliega un mensaje

# n√≥tese que no hemos especifico una respuesta en caso de que la condici√≥n no se cumpla
````

````
# vamos a expandir la prueba anterior
obj = -10 # creamos un objeto

if obj > 0:                         # condici√≥n 1: ¬øel n√∫mero es mayor a 0?
  print("number is positive")       # respuesta a condici√≥n 1
elif obj == 0:                      # condici√≥n 2: ¬øel n√∫mero es 0?
  print("number is 0")              # respuesta a condici√≥n 2
else:                               # condici√≥n 3: ninguna de las condiciones anteriores se cumple
  print("number is negative")       # respuesta a condici√≥n 3
````

### Iteraciones
Las iteraciones se articulan por medio de bucles (*loops*), estas son secuencias de instrucciones que se ejecutan un n√∫mero determinado de veces.

*¬øQu√© determina la cantidad de ejecuciones en un bucle?*

La cantidad de ejecuciones depende del tipo espec√≠fico de bucle
*   *For loop*: el bucle se repite tantas veces como la cantidad de elementos en una estructura (usualmente lists)
*   *Do while*: el bucle se ejecuta siempre y cuando una condici√≥n se mantenga cierta

Estos dos tipos de bucles son opciones t√≠picamente disponibles en cualquier lenguaje de programaci√≥n. Adicionalmente, Python incorpora una forma simplificada de realizar iteraciones: los *comprehension lists*.

#### For loops

Sintaxis b√°sica
```
my_list = [x,y,z .. n]
for i in my_list:
  print(i)
```

> üëâ El valor de nuestro objeto iterable cambia en cada ciclo

````
# ejemplo de for loop
my_list = ["a","b","c","d","e"] # creamos un list
for x in my_list: # esto se puede leer: por cada elemento x en "my_list", realizar la siguiente tarea
  print(x)

# este for loop tiene 5 iteraciones, ya que hay 5 elementos en nuestra lista
# x adquiere el valor de cada elemento en nuestra lista
````

If we want to iterate for a given large number of times (e.g. 100 times); we can use the range() function to avoid manually creating a data structure with a large number of objects stored (e.g. 100 objects).

En ocasiones, podemos querer iterar un n√∫mero espec√≠fico de veces (por ejemplo 100 veces). La funci√≥n range() nos evita el trabajo de crear manualmente un list con x elementos (ej. 100).

```
range(start,end)
```


````
# iteramos 10 veces
for value in range(1,11):
  print(value)
````

#### While loops
En un *while loop*, el bucle contin√∫a mientras una condici√≥n se cumpla.

```
while (cond):
  proceso
```

> üëâ El n√∫mero de iteraciones no est√° pre establecido, dado que no depende de la cantidad de elmentos en una estructura

> ‚ö†Ô∏è Debemos tener cuidado de evitar bucles infinitos. En tal caso, nos veremos obligados a reiniciar el entorno



````
my_object = 10                # creamos un objeto
while my_object <= 25:        # estabecemos una condici√≥n
  print(my_object)            
  my_object = my_object + 1   # sumamos 1 en cada iteraci√≥n 
````

#### Comprehension lists
